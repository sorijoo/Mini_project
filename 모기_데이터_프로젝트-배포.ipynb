### 기상청 api

```python
from pprint import pprint

from datetime import date

start = date(2020, 2, 17)
end = date(2022, 5, 27)

def days_diff(start, end):
    numOfRows = (end - start).days
    return numOfRows +1

numOfRows = days_diff(start, end)

numOfRows


url_2 = f'http://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList?serviceKey="인증키"&numOfRows={numOfRows}&pageNo=1&dataCd=ASOS&dateCd=DAY&startDt=20200217&endDt=20220525&stnIds=108&dataType=JSON'

res2 = requests.get(url_2)
data2 = res2.json()






```

```python
data2_a = data2["response"]["body"]["items"]["item"]
```

```python
data2_a[-1]
```

```python
#내가 필요한 정보로만 하루치 가져오기



keys = ['tm','avgTa','minTa','maxTa','sumRnDur','sumRn','avgWs','avgRhm','sumSsHr']



weather_list = [{k: v for k,v in data2_a[i].items() if k in keys} for i in range(len(data2_a))]

# weather_list = []
# for i in range(len(data2_a)):
#     weather_list.append({k: v for k,v in data2_a[i].items() if k in keys})
    
weather_list[1]    
```

```python
name = ['날짜','평균 기온','최저 기온','최고 기온','강수 계속시간','일강수량','평균 풍속','평균 상대습도','합계 일조 시간']
```

```python
weather_df = pd.DataFrame(weather_list)

weather_df.columns = name
weather_df = weather_df.set_index('날짜')

weather_df
```
